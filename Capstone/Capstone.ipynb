```python
# Getting average price from 2000-2023
dates = df.columns[8:]

prices = df.iloc[100, 8:]


yearly_mean_prices = []

for i in range(0, len(dates), 12):
    yearly_prices = prices[i:i+12]
    yearly_mean = np.mean(yearly_prices)
    yearly_mean_prices.append(yearly_mean)


for year, mean_price in zip(range(2000, 2024), yearly_mean_prices):
    print(f"Year {year}: Mean Price ${mean_price:,.2f}")
```

```python
import matplotlib.pyplot as plt

# Plotting the yearly mean prices
plt.figure(figsize=(12, 6))
plt.bar(range(2000, 2024), yearly_mean_prices, color='blue')
plt.xlabel('Year')
plt.ylabel('Mean Price')
plt.title('Madison Yearly Mean Housing Prices (2000-2023)')
plt.grid(True)
plt.show()
```

```python
years = list(range(2000, 2024))


plt.figure(figsize=(12, 6))
plt.plot(years, yearly_mean_prices, color='blue', marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Mean Price')
plt.title('Madison Yearly Mean Housing Prices (2000-2023)')
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("sfh.csv")
df.info()
```

```python
import pandas as pd
import numpy as np

dates = df.columns[8:]


yearly_avg_list = []

for index, row in df.iterrows():

    original_values = row.iloc[:8]
    
    prices = row[8:]
    
    yearly_avg = {"City": original_values["RegionName"],  # Change column name "RegionName" to "City"
                  "State": original_values["State"],
                  "Metro": original_values["Metro"],
                  "CountyName": original_values["CountyName"]}
    
    for i in range(0, len(dates), 12):
        yearly_prices = prices[i:i+12]
        
       
        yearly_mean = round(np.mean(yearly_prices), 2)
        
        year = int(dates[i][:4])  
        yearly_avg[str(year)] = yearly_mean
    
    yearly_avg_list.append(yearly_avg)

yearly_avg_df = pd.DataFrame(yearly_avg_list)

print(yearly_avg_df)
```

```python
yearly_avg_df = pd.DataFrame(yearly_avg_list)

csv_filename = "SingleFamilyHome.csv"
yearly_avg_df.to_csv(csv_filename, index=False)

print(f"The DataFrame has been saved to {csv_filename}")
```
